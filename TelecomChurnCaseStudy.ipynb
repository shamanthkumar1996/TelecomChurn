Handling missing values in columns

```python
# Cheking percent of missing values in columns
df_missing_columns = (round(((df.isnull().sum()/len(df.index))*100),2).to_frame('null')).sort_values('null', ascending=False)
df_missing_columns
```

```python
# List the rech columns
rec_cols = [k for k in df.columns.to_list() if 'rech' in k]
print(rec_cols) 
```

```python
df[rec_cols].describe(include='all')
```

Imputing recharge columns missing values with Zero, because if a customer had a NAN in recharge field we must have chruned.

```python
for rec in rec_cols:
    df[rec] = df[rec].fillna(0)
```

```python
# List the rech columns
rec_cols = [k for k in df.columns.to_list() if '_rech' in k]
print(rec_cols) 
```

```python
df[rec_cols].describe(include='all')
```

```python
# Count NaN values per column in reach column, we see its zero now.
nan_count = df[rec_cols].isna().sum()
print(nan_count)
```

```python
# List the columns having more than 30% missing values
col_list_missing_30 = list(df_missing_columns.index[df_missing_columns['null'] > 30])
```

```python
col_list_missing_30
```

```python
# We see lot of date columns are having high missing values and computing dates are complex
# We will drop columns with dates
# List the date columns
date_cols = [k for k in df.columns.to_list() if 'date' in k]
print(date_cols) 
# Dropping date columns
df = df.drop(date_cols, axis=1)
```

```python
df.shape
```

```python
# List the columns having more than 30% missing values
# Cheking percent of missing values in columns
df_missing_columns = (round(((df.isnull().sum()/len(df.index))*100),2).to_frame('null')).sort_values('null', ascending=False)
df_missing_columns
col_list_missing_30 = list(df_missing_columns.index[df_missing_columns['null'] > 30])
```

```python
col_list_missing_30
```

```python
# arpu is Average revenue per user 
# fb_user: Service scheme to avail services of Facebook and similar social networking sites
# night_pck_user :Scheme to use during specific night hours only
# Above mentioned data might have some meaning info especially arpu, so let us fill 0 for the rows which are empty. Considering empty might be due to churn
```
